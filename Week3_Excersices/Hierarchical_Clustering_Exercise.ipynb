#**Python Implementation using the scikit-learn library**

#Step 1 : import libraries

```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
```

#Step 2 : Randomly chosen dataset

```python
X = np.array([[1, 2], [1, 4], [1, 0],
             [4, 2], [4, 4], [4, 0]])
```

#Step 3 : Perform hierarchical clustering  + Ploting


```python
Z = linkage(X, 'ward')

#Here is the plot of dendrogram
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogran')
plt.xlabel('Data point')
plt.ylabel('Distance')
plt.show()
```

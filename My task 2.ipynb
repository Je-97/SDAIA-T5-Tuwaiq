# Tasks

```python
import pandas as pd
df1 = pd.read_csv('Task (1) Dataset.csv')
df2 = pd.read_csv('Task (2) Dataset.csv')
```

```python
column_to_merge = df1[['Weather_conditions']]
```

```python
df2 = pd.concat([df2, column_to_merge], axis=1)
```

```python
df2.head()
```

### Perform summary statistics on the dataset.

```python
df2.describe()
```

## 1. Exploratory Data Analysis (EDA)

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(df2['Datetime'], df2['TrafficIndexLive'], label='Traffic Index (Live)')
plt.plot(df2['Datetime'], df2['JamsCount'], label='Jams Count')

plt.xlabel('Datetime')
plt.ylabel('Values')
plt.title('Traffic Metrics Over Time')
plt.legend()
plt.show()

import seaborn as sns
sns.histplot(df2, x='TrafficIndexLive', bins=20, kde=True)
plt.xlabel('Traffic Index (Live)')
plt.title('Distribution of Traffic Index')
plt.show()
```

### Identify and analyze patterns in the data.

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
df2['Datetime'] = pd.to_datetime(df2['Datetime'])
df2['NumericDate'] = (df2['Datetime'] - df2['Datetime'].min()).dt.days
sns.distplot(df2['NumericDate'], kde=False, bins=20)
plt.xlabel('Numeric Date')
plt.ylabel('Frequency')
plt.title('Distribution of TrafficIndexLive over Time')
plt.show()
df2['DateCategory'] = df2['Datetime'].astype('category')
sns.countplot(x='DateCategory', data=df2)
plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Frequency')
plt.title('Distribution of TrafficIndexLive by Date')
plt.show()
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='TrafficIndexLive', y='Weather_conditions', data=df2, hue='Weather_conditions', palette='viridis', alpha=0.7)
plt.title('Relationship between Traffic Index Live and Weather Conditions', fontsize=16)
plt.xlabel('Traffic Index Live', fontsize=14)
plt.ylabel('Weather Conditions', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='Weather Conditions', fontsize=12)
plt.show()
```
